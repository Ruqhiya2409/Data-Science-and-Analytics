```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('netflix_data.csv')
df.head()
```

```python
genre_hours = df.groupby('genre')['watch_hours'].sum().sort_values(ascending=False)
print("Total watch hours by genre:\n", genre_hours)

# Bar plot
genre_hours.plot(kind='bar', color='coral', figsize=(8,6))
plt.title('Total Watch Hours by Genre')
plt.xlabel('Genre')
plt.ylabel('Total Watch Hours')
plt.show()
```

```python
user_hours = df.groupby('user_id')['watch_hours'].sum().sort_values(ascending=False)
print("Total watch hours per user:\n", user_hours)

# Bar plot
user_hours.plot(kind='bar', color='skyblue', figsize=(8,6))
plt.title('Total Watch Hours per User')
plt.xlabel('User ID')
plt.ylabel('Watch Hours')
plt.show()
```

```python
# Define binge watchers: more than 8 hours
binge_watchers = user_hours[user_hours > 8]
print("Binge watchers:\n", binge_watchers)

# Pie chart
binge_watchers.plot(kind='pie', autopct='%1.1f%%', figsize=(6,6))
plt.title('Binge Watchers Share')
plt.ylabel('')
plt.show()
```

```python
# Average rating per genre
genre_ratings = df.groupby('genre')['rating'].mean().sort_values(ascending=False)
print("Average rating per genre:\n", genre_ratings)

# Bar plot
genre_ratings.plot(kind='bar', color='lightgreen', figsize=(8,6))
plt.title('Average Rating by Genre')
plt.xlabel('Genre')
plt.ylabel('Average Rating')
plt.show()
```
