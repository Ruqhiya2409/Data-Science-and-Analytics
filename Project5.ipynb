```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('ecommerce_data.csv')
df.head()
```

```python
product_sales = df.groupby('product')['quantity'].sum().sort_values(ascending=False)

print("Top-Selling Products:\n", product_sales)

# Bar chart
product_sales.plot(kind='bar', color='skyblue', figsize=(8,6))
plt.title('Top-Selling Products')
plt.ylabel('Total Quantity Sold')
plt.show()
```

```python
df['revenue'] = df['quantity'] * df['price']
revenue = df.groupby('product')['revenue'].sum().sort_values(ascending=False)

print("Revenue by Product:\n", revenue)

# Pie chart
revenue.plot(kind='pie', autopct='%1.1f%%', figsize=(6,6))
plt.title('Revenue Share by Product')
plt.ylabel('')
plt.show()
```

```python
orders_by_hour = df.groupby('hour')['order_id'].count()

print("Orders by Hour:\n", orders_by_hour)

# Line chart
orders_by_hour.plot(kind='line', marker='o', figsize=(8,6))
plt.title('Orders by Hour')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Orders')
plt.grid(True)
plt.show()
```

```python
user_orders = df.groupby('user_id')['order_id'].count()
repeat_buyers = user_orders[user_orders > 1]

print("Repeat Buyers:\n", repeat_buyers)

# Bar chart
repeat_buyers.plot(kind='bar', color='green', figsize=(8,6))
plt.title('Repeat Buyers')
plt.xlabel('User ID')
plt.ylabel('Number of Orders')
plt.show()
```
